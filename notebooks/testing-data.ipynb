```python
import faculty.datasets as datasets
import PyPDF2
```

```python
datasets.ls()
```

```python
datasets.get('/plans2/aberdeen-city-council-5e6eab0.pdf', 'aberdeen-city-council.pdf')

pdfObj = open('aberdeen-city-council.pdf', 'rb')
reader = PyPDF2.PdfReader(pdfObj)

# Create a text file object
textFileObj = open('/project/text/aberdeen-city-council.txt', 'w')

# Loop through each page and extract text
for page in reader.pages:
  page_text = page.extract_text()
  # Write the text to the file
  textFileObj.write(page_text)

# Close both file objects
pdfObj.close()
textFileObj.close()

```

```python
import os
len(os.listdir('plans')) # 665 plans in total
pdf_files = [f for f in os.listdir('plans') if f.endswith(('.pdf','.PDF'))]
len(pdf_files)
```

```python
for f in pdf_files:
    path0 = '/project/notebooks/plans/'
    filename = path0 + f
    pdfObj = open(filename, 'rb')
    reader = PyPDF2.PdfReader(pdfObj)

    # Create a text file object
    path1 = '/project/text/'
    name = path1 + f.replace('.pdf','.txt')
    textFileObj = open(name, 'w')

    # Loop through each page and extract text
    for page in reader.pages:
      page_text = page.extract_text()
      # Write the text to the file
      textFileObj.write(page_text)

    # Close both file objects
    pdfObj.close()
    textFileObj.close()
    
```
